```python
from scipy.spatial.transform import Rotation as R
from sklearn.metrics import mean_squared_error
```

```python
# ======================================
# TRUE QUATERNIONS
# ======================================

q_true = data[['qx','qy','qz','qw']].values
```

```python
# ======================================
# PREDICTED EULER ANGLES
# ======================================

roll_pred  = y_pred[:,0]
pitch_pred = y_pred[:,1]
yaw_pred   = y_pred[:,2]
```

```python
# ======================================
# EULER -> QUATERNION
# ======================================

r = R.from_euler(
    'xyz',
    np.column_stack((roll_pred, pitch_pred, yaw_pred)),
    degrees=True
)

q_pred = r.as_quat()
```

```python
# scipy gives:
# [qx qy qz qw]

# ======================================
# MATCH TEST SIZE
# ======================================

q_true_test = q_true[-len(q_pred):]
```

```python
# ======================================
# QUATERNION RMSE
# ======================================

rmse_qx = np.sqrt(
    mean_squared_error(
        q_true_test[:,0],
        q_pred[:,0]
    )
)

rmse_qy = np.sqrt(
    mean_squared_error(
        q_true_test[:,1],
        q_pred[:,1]
    )
)

rmse_qz = np.sqrt(
    mean_squared_error(
        q_true_test[:,2],
        q_pred[:,2]
    )
)

rmse_qw = np.sqrt(
    mean_squared_error(
        q_true_test[:,3],
        q_pred[:,3]
    )
)

print("\n===== Quaternion RMSE =====")

print(f"QX RMSE : {rmse_qx:.6f}")
print(f"QY RMSE : {rmse_qy:.6f}")
print(f"QZ RMSE : {rmse_qz:.6f}")
print(f"QW RMSE : {rmse_qw:.6f}")
```

```python
# ======================================
# TOTAL QUATERNION ERROR
# ======================================

quat_error = np.mean(
    np.linalg.norm(
        q_true_test - q_pred,
        axis=1
    )
)

print(f"\nTotal Quaternion Error : {quat_error:.6f}")
```
